```python
SELECT *
FROM game_sales
ORDER BY games_sold DESC
LIMIT 10;
```

```python
SELECT g.year,COUNT(g.name) AS num_games, ROUND(AVG(critic_score),2) AS avg_critic_score
FROM game_sales AS g
INNER JOIN reviews AS r
USING(name)
GROUP BY g.year
HAVING COUNT(g.name) >=4
ORDER BY avg_critic_score DESC
LIMIT 10;
```

```python
SELECT year,u.num_games,avg_critic_score,avg_user_score,(avg_critic_score - avg_user_score) AS diff
FROM users_avg_year_rating AS u
INNER JOIN critics_avg_year_rating AS c
USING(year)
WHERE (avg_critic_score > 9 OR avg_user_score > 9)
ORDER BY year ASC
```
